# Housecare.sa home page

```
import requests
from bs4 import BeautifulSoup
import pandas as pd
web = requests.get("https://housecare.sa/")
web.content
```

```
soup = BeautifulSoup(web.content,"html.parser")
cat = soup.find_all("h3",class_ = "home-head")
```

```
item1 = soup.find_all("h2", class_ = "woocommerce-loop-product__title")
price1 = soup.find_all("span",class_ = "price")
item = []
price = []
for i in item1:
  item.append(i.text)
for j in price1:
  price.append(j.text)
ip = list(zip(item,price))
df = pd.DataFrame(data = ip, columns = ["item","price"])
df
```
